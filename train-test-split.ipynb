```python
import os
import random
import shutil
from os.path import join, basename, splitext

# Paths to the images and annotations directories
images_dir = '/Users/skshah/yolo/MAIN_IMAGE_DATASET/images'
txt_dir = '/Users/skshah/yolo/MAIN_IMAGE_DATASET/txt'

# Paths to the training and testing directories
train_images_dir = '/Users/skshah/yolo/images/train'
train_txt_dir = '/Users/skshah/yolo/labels/train'
test_images_dir = '/Users/skshah/yolo/images/test'
test_txt_dir = '/Users/skshah/yolo/labels/test'



# List all image files
image_files = [f for f in os.listdir(images_dir) if f.endswith('.png')]

# Filter out images without corresponding txt files
image_files = [f for f in image_files if os.path.exists(join(txt_dir, splitext(f)[0] + '.txt'))]

# Shuffle and split the filenames into training (70%) and testing (30%) sets
random.shuffle(image_files)
train_split = int(0.7 * len(image_files))
train_files = image_files[:train_split]
test_files = image_files[train_split:]

# Function to copy files to the respective directories
def copy_files(file_list, src_images_dir, src_txt_dir, dest_images_dir, dest_txt_dir):
    for file in file_list:
        image_file = join(src_images_dir, file)
        txt_file = join(src_txt_dir, splitext(file)[0] + '.txt')
        shutil.copy(image_file, dest_images_dir)
        shutil.copy(txt_file, dest_txt_dir)

# Copy the selected files to their respective training and testing directories
copy_files(train_files, images_dir, txt_dir, train_images_dir, train_txt_dir)
copy_files(test_files, images_dir, txt_dir, test_images_dir, test_txt_dir)

print(f"Copied {len(train_files)} files to the training set.")
print(f"Copied {len(test_files)} files to the testing set.")
```

```python
import os

# Define the directory containing the txt files
directory = '/Users/skshah/yolo/MAIN_IMAGE_DATASET/new_txt'

# Function to update class labels
def update_class_labels(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    updated_lines = []
    for line in lines:
        parts = line.strip().split()
        if parts:
            # Change class labels 0 and 1 to 2
            if parts[0] in ['0', '1']:
                parts[0] = '2'
            updated_lines.append(' '.join(parts))
    
    # Write the updated lines back to the file
    with open(file_path, 'w') as file:
        file.write('\n'.join(updated_lines) + '\n')

# Iterate over all files in the directory
for filename in os.listdir(directory):
    if filename.endswith('.txt'):
        file_path = os.path.join(directory, filename)
        update_class_labels(file_path)
        print(f'Updated {file_path}')
```
