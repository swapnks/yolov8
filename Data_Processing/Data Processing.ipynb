```python

ls = []
nls = []
path = '/Users/skshah/yolo/images/annotations/*.xml'  

for filename in glob.glob(path):
    with open(filename, 'r') as file:
        data = file.read()
    bs = BeautifulSoup(data, "xml")
    tags = bs.find_all("name")
    ntag = bs.find_all(["filename", "name"])
    nls.append(ntag)
    names = [tag.text for tag in tags]
    print(tags)
    ls.extend(names)
```

```python
## Finding out class values

s = pd.Series(ls)
print("Value counts for each class : \n",s.value_counts())
```

## copying the lowest class (mask incorrectly worn) images from the main dataset to another folder for data augmentation 

```python
source_dir = "/Users/skshah/yolo/MAIN_IMAGE_DATASET/images"
dest_dir = "/Users/skshah/yolo/images/incorrect"

incorrect_list = []
for i in nls:
    if i[1].string=="mask_weared_incorrect":
        incorrect_list.append(i[0].string)
        
print("list of images : ", incorrect_list)
```

```python
## copying from the main dataset to another folder

for file in incorrect_list:
    src_file  = os.path.join(source_dir,file)
    dest_file = os.path.join(dest_dir,file)
    try:
        shutil.copy(src_file, dest_file)
    except Exception as e:
        print(e)
    
```
